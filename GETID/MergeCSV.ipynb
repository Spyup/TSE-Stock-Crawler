```python
import os, glob
import pandas as pd
import csv
```

```python
Listed = pd.read_csv("./ListedStockID/All.csv",encoding="utf-8")
OTC = pd.read_csv("./OTCStockID/All.csv",encoding="utf-8")

merge = pd.concat([Listed,OTC],axis=0,ignore_index = True)

merge.to_csv("All.csv")
```

```python
path = "./ListedStockID/"

all_files = glob.glob(os.path.join(path,"*.csv"))
```

```python
print(all_files)
```

```python
for a in all_files:
    test = pd.read_csv(a,delimiter=",",encoding="utf-8")
    print(test)
```

```python
df_from_each_file = (pd.read_csv(f, delimiter=',',encoding="utf-8") for f in all_files)
```

```python
print(df_from_each_file)
```

```python
df_merged = pd.concat(df_from_each_file,axis=0,ignore_index = True)
#df_merged.to_csv( "merged.csv")
```

```python
print(df_merged)
```

```python
numpy_merged = df_merged.to_numpy()
```

```python
print(numpy_merged)
```

```python
with open("./ListedAll.csv",'w',newline='',encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for data_context in numpy_merged:
        writer.writerow(data_context)
```
